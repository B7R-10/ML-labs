### 1. Load the dataset into a DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_excel('global_cars_enhanced.xlsx', header=1)
df.head()
```

### 2. Explore the data (head, info, describe)

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### 3. Perform basic data cleaning if needed

```python
# Check missing values
df.isnull().sum()
```

```python
# Drop unnecessary column if exists
if 'Car_ID' in df.columns:
    df = df.drop('Car_ID', axis=1)

df.head()
```

### 4. Apply feature engineering (if applicable)

```python
# Example engineered features if they exist already
df[['HP_per_CC','Efficiency_Score','Car_Age']].head()
```

### 5. Prepare the data for modeling

```python
X = df[['Engine_CC','Horsepower','Mileage_km_per_l','Car_Age','HP_per_CC','Efficiency_Score']]
y = df['Price_USD']

X.head()
```

### 6. Train a model (Linear Regression)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=101
)

lm = LinearRegression()
lm.fit(X_train, y_train)
```

### Model Coefficients

```python
print(lm.intercept_)

coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

### 7. Evaluate the model performance

```python
predictions = lm.predict(X_test)
predictions[:10]
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
sns.histplot((y_test - predictions), bins=50)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
