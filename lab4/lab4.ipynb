Task 2: Apply one missing value strategy and explain why

```python
df_missing = df.copy()
df_missing.loc[0:5, 'Amount'] = np.nan

df_missing['Amount'] = df_missing['Amount'].fillna(df_missing['Amount'].median())

df_missing.isna().sum()
```

The median was chosen because it is less sensitive to extreme values compared to the mean.
Since the Amount column may contain outliers, the median provides a more robust replacement for missing values.

Task 3: Detect and handle outliers using IQR

```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Amount'] < lower) | (df['Amount'] > upper)]
outliers.shape
```

```python
df_no_outliers = df[(df['Amount'] >= lower) & (df['Amount'] <= upper)]

print("Original shape:", df.shape)
print("After removing outliers:", df_no_outliers.shape)
```

Task 4: Normalize numerical features using Min-Max and Z-score

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_minmax = df[['Amount', 'Boxes Shipped']].copy()

df_minmax[['Amount', 'Boxes Shipped']] = scaler.fit_transform(df_minmax)

df_minmax.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_zscore = df[['Amount', 'Boxes Shipped']].copy()

df_zscore[['Amount', 'Boxes Shipped']] = scaler.fit_transform(df_zscore)

df_zscore.head()
```

TAsk 5: Apply PCA and interpret explained variance

```python
from sklearn.decomposition import PCA

X = df_zscore[['Amount', 'Boxes Shipped']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```
